## Neural Networks para classificação do DataSet Iris

```python
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.python.keras import utils
import numpy as np
from sklearn.metrics import confusion_matrix
```

### Caregar dados e definir os Features e o Target

```python
base = datasets.load_iris()
previsores = base.data
classe = base.target
```

### Aqui aplico o One Hot Encoder no Target, uma vez que teremos 1 neurônio para classificar cada classe 

```python
classe_dummy = utils.np_utils.to_categorical(classe)
```

### Divido os Dados em Treino e Teste

```python
X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(previsores,
                                                                  classe_dummy,
                                                                  test_size = 0.3,
                                                                  random_state = 0)
```

### Iniciando a Rede Neural

```python
modelo = Sequential()
modelo.add(Dense(units = 5, input_dim = 4))
modelo.add(Dense(units = 4))
modelo.add(Dense(units = 3, activation = 'softmax'))

modelo.summary()
```

A Rede Neural possui 4 neurônios como inputs, já que utilizamos um nerônio para cada feature. Após, 5 neurônios na segunda camada, 4 na seguinte e finaliza com mais 3 neurônios, os quais, cada um irá classificar uma espécie diferente de planta

### Compilação e Treinamento da Rede

```python
modelo.compile(optimizer = 'adam', loss = 'categorical_crossentropy',
               metrics = ['accuracy'])
modelo.fit(X_treinamento, y_treinamento, epochs = 1000,
           validation_data = (X_teste, y_teste))
```

### Como o ativador softmax retorna a probabilidade, fiz o seguinte passo para ter uma resposta "binária"

```python
previsoes = modelo.predict(X_teste)
previsoes = (previsoes > 0.5)
```

### Gerando Matriz de Confusão para avaliação do modelo

```python
y_teste_matrix = [np.argmax(t) for t in y_teste]
y_previsao_matrix = [np.argmax(t) for t in previsoes]
confusao = confusion_matrix(y_teste_matrix, y_previsao_matrix)
```
